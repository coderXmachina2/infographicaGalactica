```python
#perhaps irregular Mass Bins are Ideal
#Bin 1 = Planets smaller than Earth
#Bin 2 = Earth Size
#Bin 3 = Super Earth
#Bin 4 = Jupiter Sized

bin1 = [] #Bin 1 = Terrestrial and Earth Sized
bin2 = [] #Bin 2 = superEarths
bin3 = [] #Bin 3 = Jupiter Sized
bin4 = [] #Bin 4 = superGas Giants

for p in range (0, len(validMasses)): #scan through all
    if(validMasses[p] >= 0 and  validMasses[p] < 8*0.0035): #Earth and Terrestrial Sized 0 to 8 times Eart
        bin1.append(validMasses[p])
    elif(validMasses[p] >= 8*0.0035 and validMasses[p] < 0.0035*25): #super Earths 8 times and 25 times Earth 
        #Super Earth stuffs
        bin2.append(validMasses[p])
    elif(validMasses[p] >= 0.1225 and validMasses[p] < 3):  #mid sized gas giants 
        bin3.append(validMasses[p])
    else:
        bin4.append(validMasses[p])
```
