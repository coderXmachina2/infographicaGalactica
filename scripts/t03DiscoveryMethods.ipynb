```python
import pandas as pd
import numpy as np
import sys
from matplotlib import pyplot as plt

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
exoplanets = pd.read_csv('/Users/ultra/Desktop/theJects/infographicaGalactica/data/confirmedExoplanets.csv')
data = pd.DataFrame(exoplanets)
data
```

```python
x = range(len(data))

fig = plt.figure(figsize=(20,10))
plt.scatter(x, data['pl_orbper'])
fig.suptitle('Orbital period  for each exoplanet', fontsize=14)
plt.xlabel('Exoplanet index', fontsize=14)
plt.ylabel('Orbital Period', fontsize=14)

#plt.yticks(np.arange(0, 2350, step=100))  
plt.xticks(rotation=45, fontsize=1, alpha = 0.0)

#plt.text(1200, 2.6, 'Earth orbital period', fontsize = 14)
plt.axhline(y=0, alpha= 0.5, color='blue')
#plt.axhline(y=2.55, alpha= 0.5, color='red') #earth  

#plt.axhline(y=2.8, alpha= 0.5, color='r')

plt.grid(alpha = 0.5)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/orbital_period.png')
plt.show()
```

```python
x = range(len(data))

fig = plt.figure(figsize=(20,10))
plt.scatter(x, np.log10(data['pl_orbper']))
fig.suptitle('log10 (days) Orbital period  for each exoplanet', fontsize=14)
plt.xlabel('Exoplanet index', fontsize=14)
plt.ylabel('Orbital Period log10', fontsize=14)

#plt.yticks(np.arange(0, 2350, step=100))  
plt.xticks(rotation=45, fontsize=1, alpha = 0.0)

plt.text(1200, 2.6, 'Earth orbital period', fontsize = 14)
plt.axhline(y=0, alpha= 0.5, color='blue')
plt.axhline(y=2.55, alpha= 0.5, color='red') #earth  

#plt.axhline(y=2.8, alpha= 0.5, color='r')

plt.grid(alpha = 0.5)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/orbital_period.png')
plt.show()
```

```python
sorted_period = data.sort_values(by ='pl_orbper')
df = sorted_period[sorted_period['pl_orbper'].notna()] #remove nans
x = range(len(df))
```

```python
fig = plt.figure(figsize=(20,10))
plt.scatter(x,np.log10(np.array(df['pl_orbper'][0:])))

fig.suptitle('Sorted log10 (days) Orbital period for each solar system', fontsize=14)
plt.xlabel('Exoplanet index', fontsize=14)
plt.ylabel('Orbital Period log10', fontsize=14)

#plt.yticks(np.arange(0, 2350, step=100))  
plt.xticks(rotation=45, fontsize=1, alpha = 0.0)

plt.text(0, 2.6, 'Earth orbital period', fontsize = 14)
plt.axhline(y=0, alpha= 0.5, color='blue')
plt.axhline(y=2.55, alpha= 0.5, color='red') #earth  h  
#plt.axhline(y=2.8, alpha= 0.5, color='r')

plt.grid(alpha = 0.5)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/sorted_orbital_period_log10.png')
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
plt.scatter(range(len(data)), np.log10(data['pl_orbsmax'])  )
fig.suptitle('Orbit Semi-Major Axis (AU) log 10 for each solar system', fontsize=14)
plt.xlabel('Exoplanet index', fontsize=14)
plt.ylabel('Orbit Semi-Major Axis (AU)', fontsize=14)

#plt.yticks(np.arange(0, 2350, step=100))  
plt.xticks(rotation=45, fontsize=4, alpha = 0.0001)
plt.axhline(y=0, alpha= 0.5, color='red') #1 AU 

plt.grid(alpha = 0.5)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/pl_Semi_maj_axis_AU_log10.png')
plt.show()
```
