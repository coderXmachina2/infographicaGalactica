```python
import pandas as pd
import numpy as np
import sys
from matplotlib import pyplot as plt

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
exoplanets = pd.read_csv('/Users/ultra/Desktop/theJects/infographicaGalactica/data/confirmedExoplanets.csv')
data = pd.DataFrame(exoplanets)
data
```

```python
sorted_eq_temp = data.sort_values(by ='pl_eqt') #sort
df_temp = sorted_eq_temp[sorted_eq_temp['pl_eqt'].notna()] #remove nans
x = range(len(df_temp ))
x
```

```python
fig = plt.figure(figsize=(20,10))
plt.scatter(x, df_temp['pl_eqt'])
fig.suptitle('Sorted Planet Equilibrium Temperature (K) for each exoplanet', fontsize=14)
plt.xlabel('Exoplanet index', fontsize=14)
plt.ylabel('Planet Equilibrium Temperature (k)', fontsize=14)

#plt.yticks(np.arange(0, 2350, step=100))  
plt.xticks(rotation=45, fontsize=1, alpha = 0.0)

plt.axhline(y=0, alpha= 0.5, color='blue')
plt.text(300, 260, 'Earth Planetary Equilibrium Temp', fontsize = 14)
plt.axhline(y=255, alpha= 0.5, color='red') #earth  

#plt.axhline(y=2.8, alpha= 0.5, color='r')

plt.grid(alpha = 0.5)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/planet_equilibrium_temperature_sorted.png')
plt.show()
```
