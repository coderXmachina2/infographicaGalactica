```python
import pandas as pd
import numpy as np
import astropy
from astropy import units as u
import sys
from matplotlib import pyplot as plt

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
exoplanets = pd.read_csv('/Users/ultra/Desktop/theJects/infographicaGalactica/data/confirmedExoplanets.csv')
data = pd.DataFrame(exoplanets)
data
```

```python
print("Since 1989 a total of: ", (len(data)), " exoplanets have been discovered")
```

```python
#discovery method
unique_disc_method, count_disc_method = np.unique(np.array(data['pl_discmethod']), return_counts=True) 
dict(zip(unique_disc_method, count_disc_method))
```

```python
fig = plt.figure(figsize=(20,10))
plt.bar(unique_disc_method, count_disc_method )
fig.suptitle('Planet Discoveries vs methods', fontsize=14)
plt.xlabel('Method', fontsize=14)
plt.ylabel('N exoplanets', fontsize=14)
#plt.yticks(np.arange(0, 110, step=10))
plt.grid(alpha = 0.4)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/Methods_of_discovery.png')
#plt.show()
```

```python
fig1, ax1 = plt.subplots()

tab =unique_disc_method.tolist()

a = []
for items in range(0, len(count_disc_method)):
    if(count_disc_method[items] > 40):
        a.append(unique_disc_method.tolist()[items])        
    else:
        a.append('')
        
ax1.pie(count_disc_method.tolist(),
        explode = (0, 0, 0, 0.01, 0, 0, 0, 0.01, 0.01, 0),
        labels=a,
        autopct='%1.2f%%',
        shadow=False,  )

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.suptitle('Most effective planet hunting methods', fontsize=14)
plt.savefig('/Users/ultra/Desktop/theJects/infographicaGalactica/Vissies/pie_Methods_of_discovery.png')
plt.show()
```
