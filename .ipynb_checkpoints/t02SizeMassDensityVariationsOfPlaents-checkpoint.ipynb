```python
#Visualizing 2nd Set of Data
#t02 Variations in mass/size of exoplanet and densities

import pandas as pd
import numpy as np
import sys
from matplotlib import pyplot as plt

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
#the first graph displays the rate of discovery of exoplanet
dFrame = pd.read_csv('confirmedExoplanets.csv')
exoRecArr  = dFrame.to_records(index=False)
confirmedExo = np.array(exoRecArr.tolist()) #either a dont delete it or adaptfurther down.
```

```python
#we have verified some crucial characteristics

# 11 is the mass of the world
# 13 is the radius

confirmedExo[0]
```

```python
#pull me up mass. Mass would be jupiter mass
#be advised Mass has missing values
#np.nanmin(confirmedExo[:,11].astype(np.float))
#np.nanmax(confirmedExo[:,11].astype(np.float))
#print("Checker: " + str(confirmedExo[0][11]))

smallestPlanet = '{0:.10f}'.format(np.nanmin(confirmedExo[:,11].astype(np.float)))
largestPlanet = '{0:.10f}'.format(np.nanmax(confirmedExo[:,11].astype(np.float)))
earthJMass = 0.000315

print("The smallest world we've discovered " + str(  '{0:.10f}'.format(np.nanmin(confirmedExo[:,11].astype(np.float)))  ) + " Jupiter Masses" )#11 is the mass
print("The largest world we've discovered " + str(  '{0:.10f}'.format(np.nanmax(confirmedExo[:,11].astype(np.float)))  ) + " Jupiter Masses" )#11 is the mass

print("\nPlanet Earth is 0.000315 Jupiter Mass")

#small array

a = [ smallestPlanet , float(earthJMass) , float(1) , largestPlanet  ]

a[0] = float(a[0])
a[1] = float(a[1])
a[3] = float(a[3])
```

```python
print(type(a[0]))
print(type(a[1]))
print(type(a[2]))
print(type(a[3]))
```

```python
#smallest world ever discovered was via pulsar timing if im not mistaken

for t in range(0, len(confirmedExo)):
    if( confirmedExo[t][11].astype(np.float) == np.nanmin(confirmedExo[:,11].astype(np.float))      ):
        print(confirmedExo[t][3]) # bingo

for t in range(0, len(confirmedExo)):
    if( confirmedExo[t][11].astype(np.float) == np.nanmax(confirmedExo[:,11].astype(np.float))      ):
        print(confirmedExo[t][3]) # bingo
```
